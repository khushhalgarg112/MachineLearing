```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
frame = cv2.imread("Personal Image.jpeg")

# Cv2 read it as BGR not RGB which maek colors invert
```

```python
plt.imshow(frame)
```

```python
#  to make it original we convert he color

rgb = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)

plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("models/haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(frame)
```

```python
face = faces[0]
```

```python
x,y,w,h = face
```

```python
face = rgb[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(rgb,(x,y), (x+w,y+h), (0,255,0),8)
plt.imshow(out)
```
