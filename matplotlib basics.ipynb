```python
# Plotting a line
x = np.linspace(1,20,1000)
```

```python
m,c = 2,10
```

```python
y = m*x + c
```

```python
# plt.plot(x,y) Plot the line
plt.scatter(x,y)  # Represent Point rather then plotting the line
```

```python
noise = np.random.randn(1000) *5
```

```python
y_mod = y + noise
```

```python
plt.scatter(x,y_mod)
```
